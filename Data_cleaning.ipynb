```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('startup_funding.csv')
```

```python
df
```

```python
df.info()
```

```python
df['Industry Vertical']=df['Industry Vertical'].replace('\\\\x','',regex=True)
df['Industry Vertical']=df['Industry Vertical'].replace('xa0','',regex=True)
df['Industry Vertical']=df['Industry Vertical'].replace('xc2','',regex=True)
```

```python
df['InvestmentnType'].unique()
```

```python
df['InvestmentnType']=df['InvestmentnType'].replace('\\\\\\n','',regex=True)
```

```python
df['InvestmentnType'].unique()
```

```python
df['Amount in USD']=df['Amount in USD'].fillna('0')
```

```python
df['Amount in USD']
```

```python
df.info()
```

```python
df.drop('Remarks', axis=1,inplace=True)
```

```python
df.set_index('Sr No',inplace=True)
```

```python
df.rename(columns={
    'Date dd/mm/yyyy':'Date',
    'Startup Name':'startup',
    'Industry Vertical':'vertical',
    'SubVertical':'subvertical',
    'City  Location':'city',
    'Investors Name':'investors',
    'InvestmentnType':'round',
    'Amount in USD':'amount'
},inplace=True)
```

```python
df.info()
```

```python
df['amount']=df['amount'].str.replace(',','')
df['amount']=df['amount'].replace('undisclosed','0')
df['amount']=df['amount'].replace('unknown','0')
df['amount']=df['amount'].replace('Undisclosed','0')
df=df[df['amount'].str.isdigit()]
df['amount']=df['amount'].astype(float)
```

```python
df.shape
```

```python
def to_inr(dollar):
    inr=dollar*87.43
    return inr/10000000
```

```python
df['amount']=df['amount'].apply(to_inr)
```

```python
df['Date']=df['Date'].str.replace('05/072018','05/07/2018')
```

```python
df['Date']=pd.to_datetime(df['Date'],format="%d/%m/%Y",errors='coerce')
```

```python
df.info()
```

```python
df=df.dropna(subset=['Date','startup','vertical','city','investors','round','amount'])
```

```python
df.info()
```

```python
df
```

```python
df.to_csv('Startup_cleaned.csv')
```
